```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Cargar la imagen y visualizarla en formato RGB correcto
def cargar_imagen_rgb(ruta_imagen):
    # Cargar imagen con OpenCV (en formato BGR)
    imagen_bgr = cv2.imread(ruta_imagen)
    # Convertir de BGR a RGB para visualizacin correcta
    imagen_rgb = cv2.cvtColor(imagen_bgr, cv2.COLOR_BGR2RGB)
    return imagen_rgb, imagen_bgr

# 2. Convertir a espacio de color HSV
def convertir_hsv(imagen_rgb):
    return cv2.cvtColor(imagen_rgb, cv2.COLOR_RGB2HSV)

# 3. Visualizar las bandas R, G, B, H, S, V
def visualizar_bandas(imagen_rgb, imagen_hsv):
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))

    # Bandas RGB
    axes[0, 0].imshow(imagen_rgb[:, :, 0], cmap='Reds')
    axes[0, 0].set_title('Canal R')
    axes[0, 1].imshow(imagen_rgb[:, :, 1], cmap='Greens')
    axes[0, 1].set_title('Canal G')
    axes[0, 2].imshow(imagen_rgb[:, :, 2], cmap='Blues')
    axes[0, 2].set_title('Canal B')

    # Bandas HSV
    axes[1, 0].imshow(imagen_hsv[:, :, 0], cmap='hsv')
    axes[1, 0].set_title('Canal H (Hue)')
    axes[1, 1].imshow(imagen_hsv[:, :, 1], cmap='gray')
    axes[1, 1].set_title('Canal S (Saturation)')
    axes[1, 2].imshow(imagen_hsv[:, :, 2], cmap='gray')
    axes[1, 2].set_title('Canal V (Value)')

    plt.tight_layout()
    plt.show()

# 4. Segmentacin por color - Experimentar con umbrales
def segmentar_senal_ciclovia(imagen_hsv):
    """
    Segmenta la seal de ciclova basndose en colores caractersticos.
    Para una seal de ciclova tpica (azul/blanca), usamos umbrales en HSV.
    """
    # Umbrales para el color azul (comn en seales de trnsito)
    # Estos valores pueden necesitar ajuste segn la imagen especfica

    # Rango para azules (Hue entre 100-130 en escala 0-180 de OpenCV)
    lower_blue = np.array([100, 50, 50])
    upper_blue = np.array([130, 255, 255])

    # Rango para blancos/grises (baja saturacin, alto valor)
    lower_white = np.array([0, 0, 200])
    upper_white = np.array([180, 50, 255])

    # Crear mscaras
    mask_blue = cv2.inRange(imagen_hsv, lower_blue, upper_blue)
    mask_white = cv2.inRange(imagen_hsv, lower_white, upper_white)

    # Combinar mscaras (seal puede tener partes azules y blancas)
    mask_combined = cv2.bitwise_or(mask_blue, mask_white)

    # Operaciones morfolgicas para limpiar la mscara
    kernel = np.ones((3, 3), np.uint8)
    mask_cleaned = cv2.morphologyEx(mask_combined, cv2.MORPH_OPEN, kernel)
    mask_cleaned = cv2.morphologyEx(mask_cleaned, cv2.MORPH_CLOSE, kernel)

    return mask_cleaned, mask_blue, mask_white

# 5. Visualizar segmentacin
def visualizar_segmentacion(imagen_rgb, mascara):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    # Imagen original
    axes[0].imshow(imagen_rgb)
    axes[0].set_title('Imagen Original RGB')
    axes[0].axis('off')

    # Mscara de segmentacin
    axes[1].imshow(mascara, cmap='gray')
    axes[1].set_title('Mscara de Segmentacin')
    axes[1].axis('off')

    plt.tight_layout()
    plt.show()

    # Opcional: Mostrar solo los objetos segmentados en color
    imagen_segmentada = cv2.bitwise_and(imagen_rgb, imagen_rgb, mask=mascara)

    plt.figure(figsize=(10, 5))
    plt.imshow(imagen_segmentada)
    plt.title('Objetos Segmentados en Color')
    plt.axis('off')
    plt.show()

# 7. Aplicar operador Canny
def aplicar_canny(mascara, umbral_bajo=50, umbral_alto=150):
    """
    Aplica el operador Canny para deteccin de bordes
    """
    bordes = cv2.Canny(mascara, umbral_bajo, umbral_alto)
    return bordes

# 8. Visualizar resultados completos
def visualizar_completo(imagen_rgb, mascara_segmentacion, bordes_canny):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    # Imagen original RGB
    axes[0].imshow(imagen_rgb)
    axes[0].set_title('Imagen Original')
    axes[0].axis('off')

    # Mscara de segmentacin
    axes[1].imshow(mascara_segmentacion, cmap='gray')
    axes[1].set_title('Segmentacin por Color')
    axes[1].axis('off')

    # Bordes de Canny
    axes[2].imshow(bordes_canny, cmap='gray')
    axes[2].set_title('Bordes (Canny)')
    axes[2].axis('off')

    plt.tight_layout()
    plt.show()

# Funcin principal
def procesar_imagen_seal(ruta_imagen):
    # 1. Cargar imagen
    imagen_rgb, _ = cargar_imagen_rgb(ruta_imagen)

    # 2. Convertir a HSV
    imagen_hsv = convertir_hsv(imagen_rgb)

    # 3. Visualizar bandas
    print("Visualizando bandas de color...")
    visualizar_bandas(imagen_rgb, imagen_hsv)

    # 4. Segmentar seal de ciclova
    print("Segmentando seal de ciclova...")
    mascara_segmentacion, mask_blue, mask_white = segmentar_senal_ciclovia(imagen_hsv)

    # 5. Visualizar segmentacin
    print("Visualizando resultados de segmentacin...")
    visualizar_segmentacion(imagen_rgb, mascara_segmentacion)

    # 6. Explicar criterio de segmentacin
    print("\n--- CRITERIO DE SEGMENTACIN ---")
    print("Se utilizaron los siguientes criterios:")
    print("- Color AZUL: Hue 100-130, Saturacin 50-255, Valor 50-255")
    print("- Color BLANCO: Hue 0-180, Saturacin 0-50, Valor 200-255")
    print("Se combinaron ambas mscaras para capturar toda la seal")
    print("Se aplicaron operaciones morfolgicas para reducir ruido")

    # 7. Aplicar Canny con diferentes parmetros
    print("\nProbando diferentes parmetros de Canny...")

    # Probar diferentes umbrales
    umbrales_canny = [
        (30, 100),   # Bajo contraste
        (50, 150),   # Medio
        (100, 200)   # Alto contraste
    ]

    for i, (umbral_bajo, umbral_alto) in enumerate(umbrales_canny):
        bordes = aplicar_canny(mascara_segmentacion, umbral_bajo, umbral_alto)

        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        plt.imshow(mascara_segmentacion, cmap='gray')
        plt.title('Mscara Original')
        plt.axis('off')

        plt.subplot(1, 2, 2)
        plt.imshow(bordes, cmap='gray')
        plt.title(f'Canny: ({umbral_bajo}, {umbral_alto})')
        plt.axis('off')

        plt.tight_layout()
        plt.show()

    # 8. Visualizacin final completa
    print("\nVisualizacin final completa...")
    bordes_finales = aplicar_canny(mascara_segmentacion, 50, 150)
    visualizar_completo(imagen_rgb, mascara_segmentacion, bordes_finales)

    # 9. Pregunta de reflexin
    print("\n--- REFLEXIN SOBRE EL OPERADOR CANNY ---")
    print("Qu efecto tiene modificar el umbral del operador Canny?")
    print("""
    Efectos de modificar los umbrales de Canny:

    1. UMBRAL BAJO (ej: 30-100):
       - Detecta ms bordes, incluyendo los dbiles
       - Mayor sensibilidad al ruido
       - Puede generar bordes falsos o discontinuos

    2. UMBRAL MEDIO (ej: 50-150):
       - Balance entre deteccin de bordes y reduccin de ruido
       - Detecta bordes significativos manteniendo buena continuidad

    3. UMBRAL ALTO (ej: 100-200):
       - Detecta solo bordes muy fuertes y contrastados
       - Menor sensibilidad al ruido
       - Puede perder bordes dbiles pero importantes

    La eleccin depende del objetivo: para seales bien definidas,
    umbrales medios-altos suelen funcionar mejor.
    """)

    return imagen_rgb, imagen_hsv, mascara_segmentacion, bordes_finales

# Ejecutar el procesamiento completo
if __name__ == "__main__":
    # Reemplaza con la ruta correcta de tu imagen
    ruta_imagen = "/content/Copia de imagen_ejercicio_3.jpg"

    try:
        imagen_rgb, imagen_hsv, mascara, bordes = procesar_imagen_seal(ruta_imagen)
        print("Procesamiento completado exitosamente!")

    except Exception as e:
        print(f"Error al procesar la imagen: {e}")
        print("Asegrate de que la ruta de la imagen es correcta")
```
